```python
import pandas as pd

#read in the two parts, Medicaid and CHIP FTE
df_old = pd.read_csv('BOD_final_0.csv')
df_new = pd.read_csv('BOD_final.csv')
```

```python
No_State = df_new[(df_new.BOD_Practice_State == '-')]
```

```python
No_State.BOD_Practice_State.value_counts()
```

```python
Yes_State = df_old[(df_old.BOD_Practice_State != '-')]
```

```python
Yes_State_Name_List = Yes_State.BOD_formatted_name.unique().tolist()
```

```python
No_State_Name_List = No_State.BOD_formatted_name.unique().tolist()
```

```python
df_yes_total = pd.DataFrame()

for bRow in Yes_State.itertuples():
    if bRow.BOD_formatted_name in No_State_Name_List:  #pick out the names which are not in NPPRES
            df_yes_total = df_yes_total.append({
                'BOD_Last_Name':bRow.BOD_Last_Name, 
                'BOD_First_Name':bRow.BOD_First_Name, 'BOD_Middle_Name':bRow.BOD_Middle_Name, 
                'BOD_lic_id':bRow.BOD_lic_id, 'BOD_Expire_Date':bRow.BOD_Expire_Date, 'BOD_Lic_Activity':bRow.BOD_Lic_Activity,  
                'BOD_Modifier_Descrip':bRow.BOD_Modifier_Descrip,'BOD_Modifier_Type':bRow.BOD_Modifier_Type, 'BOD_Practice_Name':bRow.BOD_Practice_Name, 
               'BOD_Practice_location1':bRow.BOD_Practice_location1,'BOD_Practice_Location2':bRow.BOD_Practice_Location2,'BOD_Practice_City':bRow.BOD_Practice_City,
                  'BOD_Practice_zip':bRow.BOD_Practice_zip,
                'BOD_Practice_State':bRow.BOD_Practice_State, 'BOD_Practice_county':bRow.BOD_Practice_county,'BOD_names':bRow.BOD_names,
                'BOD_formatted_name':bRow.BOD_formatted_name,'BOD_addr':bRow.BOD_addr
            },ignore_index=True)
                    
```

```python
df_yes_total
```

```python
df_yes_matched_names = df_yes_total.BOD_formatted_name.unique().tolist()
```

```python
df_No_State_remove_names = df_new[~df_new['BOD_formatted_name'].isin(df_yes_total['BOD_formatted_name'])]
```

```python
df_No_State_remove_names.shape
```

```python
df_main= pd.concat([df_No_State_remove_names, df_yes_total])
```
